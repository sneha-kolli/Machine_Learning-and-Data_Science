```python
#let's create a data frame all by ourselves.

data1 = {'S':[1,2,3,4],'N':[5,6,7,8],'E':[9,10,11,12],'H':[13,14,15,16],'A':[17,18,19,20]}
```

```python
# to fetch the column values
df2.E
```

```python
df2.A
```

loc is reffered as local indexing where as iloc is used for integer indexing.

```python
df2.iloc[3][2] # it resembles 3rd row and 2nd column
```

```python
df2.iloc[3,2] = 24 # remember that the indexing basically starts from zero and goes on
```

```python
df2
```

```python
# if you want to add a new row to insert new data you can try the following

df2.loc[4] = [5,33,25,54,68]
```

```python
df2.loc[10] = [27,55,38,90,180]
```

```python
df2
```

```python
# to update a single value you can make use of iloc
df2.iloc[0,3] = 133
```

```python
df2
```

```python
# to remove the column we can make use of drop but it won't store the data  in the dataframe
df2.drop('A',axis = 1)
# axis = 0 => for row
# axis = 1 => for column
```

```python
df2 # but here you can see that the data is not manipulated in the data frame so we need to do some more computations.
```

```python
df2.drop('A',axis = 1 , inplace = True) # inplace is a keyword which make sures that the dataframe remains consistent with all the changes.
```

```python
df2
```

```python
# you can sort the data frame based on the column or the row
```

```python
df2.reset_index(drop = True,inplace = True) # to get back into the original format.
# if you want to maintain the consistency you can add inplace = True
```

```python
df2
```

```python
df2.sort_index(ascending=True)
```

```python
df2.sort_index(ascending=False)
```

```python
# to sort the column
df2.sort_values(by='S',ascending = True)
```

```python
df2.sort_values(by='S',ascending = False)
```
