```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.linspace(0,10,11)
x
```

The numpy. linspace() function is used to create an array of evenly spaced numbers within a specified range.

```python
sin_ = np.sin(x)
sin_
```

```python
cos_ = np.cos(x)
cos_
```

```python
plt.plot(x,sin_)
```

```python
plt.plot(x,cos_)
```

```python
plt.scatter(x,sin_)
```

```python
plt.scatter(x,cos_)
```

```python
x = np.linspace(0,10,50)
x
```

```python
sin_ = np.sin(x)
sin_
```

```python
plt.plot(x,sin_)
```

```python
plt.scatter(x,sin_)
```

```python
plt.figure(figsize = (2,2)) # it is basically used to create a figure object with figsize = 2,2
plt.scatter(x,sin_)
```

```python
plt.figure(figsize = (5,2)) # it is basically used to create a figure object with figsize = 5,2
plt.scatter(x,sin_)
```

```python
plt.figure(figsize = (15,9))
plt.xlabel("linear values")
plt.ylabel("sin values")
plt.title("Sin Graph",fontsize = 15)
plt.scatter(x,sin_)
plt.show()
```

```python
plt.figure(figsize = (15,9))
plt.xlabel("linear values")
plt.ylabel("sin values")
plt.title("Sin Graph",fontsize = 15)
plt.plot(x,sin_,c="b") # c represents color
plt.show()
```

```python
plt.figure(figsize = (15,9))
plt.xlabel("linear values")
plt.ylabel("sin values")
plt.title("Sin Graph",fontsize = 15)
plt.plot(x,sin_,marker = 'd',markersize=12) # marker basically represents the locus representation shape of that specific point
plt.show()
```

```python
plt.figure(figsize = (15,9))
plt.xlabel("linear values")
plt.ylabel("sin values")
plt.title("Sin Graph",fontsize = 15)
plt.plot(x,sin_,marker = 'd',markersize=12,linestyle = "dashed") # marker basically represents the locus representation shape of that specific point
plt.show()
```

Let's plot both sin and cos plots

```python
plt.figure(figsize = (15,9))
plt.xlabel("linear_values")
plt.ylabel("trigo_values")
plt.title("trigo graphs")
plt.plot(x,sin_,marker="o",c = "red")
plt.scatter(x,np.cos(x),marker = "d",c = "green")
plt.show()
```

for giving your own axial numericals we can opt for ticks.

```python
plt.figure(figsize = (15,9))
plt.xlabel("linear_values")
plt.ylabel("trigo_values")
plt.title("trigo graphs")
plt.plot(x,sin_,marker="o",c = "red")
plt.scatter(x,np.cos(x),marker = "d",c = "green")
plt.xticks([2,4,6,8,10])
plt.show()
```

```python
plt.figure(figsize = (15,9))
plt.xlabel("linear_values")
plt.ylabel("trigo_values")
plt.title("trigo graphs")
plt.plot(x,sin_,marker="o",c = "red")
plt.scatter(x,np.cos(x),marker = "d",c = "green")
plt.xticks([1,3,5,7,9])
plt.show()
```

```python
plt.figure(figsize = (15,9))
plt.xlabel("linear_values")
plt.ylabel("trigo_values")
plt.title("trigo graphs")
plt.plot(x,sin_,marker="o",c = "red")
plt.scatter(x,np.cos(x),marker = "d",c = "green")
plt.xticks(range(1,11,3))
plt.show()
```

similarly to work it out for y we can use of yticks and to limit them we can make use of xlim and ylim

```python
plt.figure(figsize = (15,9))
plt.xlabel("linear_values")
plt.ylabel("trigo_values")
plt.title("trigo graphs")
plt.plot(x,sin_,marker="o",c = "red")
plt.scatter(x,np.cos(x),marker = "d",c = "green")
plt.xticks(range(1,11,3))
plt.xlim([0,2*3.14])
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (15,9))
plt.xlabel("linear_values")
plt.ylabel("trigo_values")
plt.title("trigo graphs")
plt.plot(x,sin_,marker="o",c = "red")
plt.scatter(x,np.cos(x),marker = "d",c = "green")
plt.xticks(range(1,11,3))
plt.grid() # to add grid to the graph or visualization drawn
plt.show()
```

```python
plt.figure(figsize = (15,9))
plt.xlabel("linear_values")
plt.ylabel("trigo_values")
plt.title("trigo graphs")
plt.plot(x,sin_,marker="o",c = "red",label = "sin")
plt.scatter(x,np.cos(x),marker = "d",c = "green",label  ="cos")
plt.xticks(range(1,11,3))
plt.grid() # to add grid to the graph or visualization drawn
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (15,9))
plt.xlabel("linear_values")
plt.ylabel("trigo_values")
plt.title("trigo graphs")
plt.plot(x,sin_,marker="o",c = "red")
plt.plot([0,100],[0,100])
plt.scatter(x,np.cos(x),marker = "d",c = "green")
plt.xticks(range(1,11,3))
plt.grid() # to add grid to the graph or visualization drawn
plt.show()
```

```python
plt.figure(figsize = (15,9))
plt.xlabel("linear_values")
plt.ylabel("trigo_values")
plt.title("trigo graphs")
plt.plot(x,sin_,marker="o",c = "red")
plt.plot([0,10],[0,10])
plt.scatter(x,np.cos(x),marker = "d",c = "green")
plt.xticks(range(1,11,3))
plt.grid() # to add grid to the graph or visualization drawn
plt.show()
```
